```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Empty list
data = []
```

```python
#For Beds
category_url = 'https://www.ikea.com/sa/en/cat/beds-bm003/'
#Set Number of pages
num_of_pages = 5
#Set Number of products
num_of_products = 100
products_scraped = 0

for i in range(1, num_of_pages+1):
    resp = requests.get(f'{category_url}?page={num_of_pages}')
    soup = BeautifulSoup(resp.text, 'html.parser')

    products = soup.select('.plp-fragment-wrapper')
    contains_products = int((((soup.select('[aria-live="assertive"]')[0].getText()).split('of'))[-1]).strip())
    print(f'Found {len(products)} products on the specified url, whereas this category contains {contains_products} products')
    
    category = soup.select('.plp-page-title__title')[0].getText()
    
    for j in range(len(products)):
        if products_scraped >= num_of_products:
            break
            
        try:
            product_name = products[j].select('.range-revamp-header-section__title--small')[0].getText()
        except:
            product_name = ''
            
        try:
            product_price = products[j].select('.range-revamp-compact-price-package__price-wrapper')[0].getText()
        except:
            product_price = ''
            
        try:
            product_description = products[j].select('.range-revamp-header-section__description-text')[0].getText()
        except:
            product_description = ''
            
        try:
            product_measurement = products[j].select('.range-revamp-header-section__description-measurement')[0].getText()
        except:
            product_measurement = ''
        
        data.append([product_name, product_price, product_description, product_measurement, category])
        print(data[-1], len(data))
        products_scraped += 1
```
